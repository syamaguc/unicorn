```python
# 2021-2025 東京開催/芝/マイル
df = result_df("2005", "2025", ["東"])
cond = df["確定着順"] == 1
df = df[cond]
df_result = df[
    [
        "レースID",
        "馬名",
        "場所",
        "芝・ダ",
        "距離",
        "クラスコード",
        "馬場状態",
        "レース日",
        "走破タイム",
    ]
]
df_result.loc[:, "レースID"] = df_result["レースID"].astype(str)
mask = df_result["レースID"].str.len() == 9
# 先頭に "0" を付けて10桁に
df_result.loc[mask, "レースID"] = df_result.loc[mask, "レースID"].str.zfill(10)
# 馬番を削る
df_result.loc[:, "レースID"] = df_result["レースID"].str[:-2]

## 2021-2025 東京開催
df_time = time_df("2005", "2025", ["東"])
df_time["レースID"] = df_time["レースID"].astype(str)
mask = df_time["レースID"].str.len() == 7
# 先頭に "0" を付けて8桁に
df_time.loc[mask, "レースID"] = df_time.loc[mask, "レースID"].str.zfill(8)
```

```python
df_joined = df_time.merge(
    df_result,
    how="left",
    on="レースID",
    suffixes=("", "_r"),
)
```
