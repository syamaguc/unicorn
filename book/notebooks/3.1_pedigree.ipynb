```python
df = result_df("2024", "2024")
```

```python
# とりあえずはサンプルの多い、父TOP10か母父TOP10から
x = 10

# ------------------------------------------------------------------
# 1. 父
# ------------------------------------------------------------------
top_father = (
    df.groupby("父馬名")
    .size()  # 件数を数える
    .reset_index(name="count")  # DataFrame に戻す
    .sort_values("count", ascending=False)  # 降順ソート
    .head(x)  # 上位 x 件だけ取り出す
)

# ------------------------------------------------------------------
# 2. 母父
# ------------------------------------------------------------------
top_grand_father = (
    df.groupby("母の父馬名")
    .size()
    .reset_index(name="count")
    .sort_values("count", ascending=False)
    .head(x)
)


# ------------------------------------------------------------------
# 3. 父 x 母父
# ------------------------------------------------------------------
top_multi = (
    df.groupby(["父馬名", "母の父馬名"])
    .size()
    .reset_index(name="count")
    .sort_values("count", ascending=False)
    .head(x)
)
```
